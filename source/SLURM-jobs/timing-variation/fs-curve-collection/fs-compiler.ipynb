```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
for timing_offset in range(20):

    f_vec = np.zeros((1, 41), dtype=float)
    for trial in range(1, 101):
        filename = "fs-collection-timing-raw-shorter-tau/offset-" + str(timing_offset) + "-trial-" + str(trial) + ".csv"
        data = np.loadtxt(filename, delimiter=",")
        f = data
        
        if trial > 1:
            f_vec = np.vstack([f_vec, np.zeros((1, 41), dtype=float)])

        f_vec[-1] = f

    f_vec = f_vec.mean(axis=0)

    new_filename = "fs-collection-timing-compiled-shorter-tau/timing-offset-fs-" + str(timing_offset) + ".csv"
    np.savetxt(new_filename, f_vec, delimiter=",")

    plot_filename = "fs-collection-timing-plots-shorter-tau/timing-offset-fs-plot-" + str(timing_offset) + ".svg"
    plt.rcParams['xtick.labelsize'] = 14
    plt.rcParams['ytick.labelsize'] = 14
    fig, axs = plt.subplots(figsize=(5,5))
    axs.plot(f_vec)
    axs.set_xlabel('Input (# of Synapses)', fontsize=14)
    axs.set_ylabel('Response (Hz)', fontsize=14)
    title_string = 'FS Curve ($∂_{Excit}$: ' + str(timing_offset) + ' ms)'
    axs.set_title(title_string, fontsize=14)
    axs.grid(visible=True, which='major')
    fig.savefig(plot_filename)

```
