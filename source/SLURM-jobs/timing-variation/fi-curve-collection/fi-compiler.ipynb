```python
import numpy as np
from matplotlib import pyplot as plt
```

```python


f_vec = np.zeros((1, 61), dtype=float)
for trial in range(1, 101):
    filename = "fi-collection-timing/trial-" + str(trial) + ".csv"
    data = np.loadtxt(filename, delimiter=",")
    inj_i = data[0]
    f = data[1]
    
    if trial > 1:
        f_vec = np.vstack([f_vec, np.zeros((1, 61), dtype=float)])

    f_vec[-1] = f

print(np.size(f_vec))
f_vec = f_vec.mean(axis=0)
new_filename = "fi-timing.csv"
np.savetxt(new_filename, np.vstack((inj_i, f_vec)).T, delimiter=",")

plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
fig, axs = plt.subplots(figsize=(5,5))
axs.plot(inj_i, f_vec)
axs.set_xlabel('Current Input (nA)', fontsize=14)
axs.set_ylabel('Response (Hz)', fontsize=14)
axs.set_title('Somatic FI Curve', fontsize=14)
axs.grid(visible=True, which='major')
fig.savefig("fi-timing-plot.svg")
```
