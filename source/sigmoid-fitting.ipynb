```python
# IMPORTS

import random
import csv
import scipy
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation
from tqdm.notebook import tqdm

from time import time
import os.path

from socket import gethostname
from scipy.optimize import curve_fit

%config InlineBackend.figure_formats = ['svg']
```

```python
def sigmoid(x, x0, B, k=1):
    """
    x: x data point
    x0: middle x data point of the sigmoid (inflection point)
    A: scaling output factor (y: [0, A]) (typically, y_max)
    B: horizontal shift (typically, y_min)
    k: input scale (default to 1)
    """
    A = yS_max
    y = A / (1 + np.exp(-k*(x-x0))) + B
    
    return y

def sigmoid_fit(x_data, y_data, p0):
    """
    x_data: array of x_data points (# of synapses)
    y_data: array of y_data points (frequency or current at soma)
    p0: initial parameters for sigmoid fit
    
    Generates a sigmoidal fit for a range of data
    """
    
    # Adds weight to the first data point such that the fit is forced into the first data point
    len_y = len(y_data)
    weights = np.ones(len_y)
    weights[0] = 0.001
    y_input = y_data
    
    # Generates sigmoidal fit
    popt, pcov = curve_fit(sigmoid, x_data, y_input, p0, method='dogbox', sigma=weights, maxfev=100000)
    return popt, pcov

def sigmoid_plot(x_data, p0):
    """
    x_data: typically # of synapses
    p0: curve parameters
    """
    
    # plots sigmoidal data
    y = sigmoid(x_data, p0[0], p0[1], p0[2])
    plt.plot(x_data, y)
    
def line_plot(x_data, p, projection=False):
    """
    x_data: typically # of synapses
    p: line parameters (m, b)
    """
    
    # plots line
    y = p[0] * x_data + p[1]
    if projection:
        plt.plot(x_data, y, '--')
    else:
        plt.plot(x_data, y)
    
def rmse(predictions, targets):
    """
    predictions: data points generated from fit
    targets: actual data points
    
    Calculates root mean square error for each data point
    """
    return np.sqrt(((predictions - targets) ** 2).mean())
    
def fs_fit(syns, y):
    """
    syns: vector of # of synapses
    y: y data points (can be frequency or current output at soma)
    
    
    For a given FS or IS curve, fits a linear segment in the beginning and a sigmoidal segment thereafter
    Iterates through multiple configurations and using the one with the best fit
    """
    
    # Synpase value at which the fit switches from linear to sigmoidal
    transition_syn = 8
    
    # Minimum error (defaulted to 1000)
    min_error = 1000
    
    # Best transition synapse value (what we are looking for
    best_ts = 0
    
    # iterates through all possible transition points between linear/sigmoid fit
    for t_s in range(2, 10):
        
        
        transition_syn = t_s
        
        # Grab linear portion of data
        syns_L = syns[:transition_syn+1]
        y_L = y[:transition_syn+1]
    
        # Grab sigmoid portion of data
        syns_S = syns[transition_syn:] 
        y_S = y[transition_syn:]

        # Perform fits on each:
        
        ## LINEAR PORTION
        p = np.polyfit(syns_L, y_L, 1)

        ## SIGMOID PORTION
        y_S[0] = p[0] * transition_syn + p[1] # FORCE first data point to be attached to end of linear fit
        
        # p0 = [x0, A, B, k]
        p0 = [np.median(syns_S), min(y_S), 1]
        popt, pcov = sigmoid_fit(syns_S, y_S, p0)
        
        # Calculate weighted RMSE, smallest is best fit
        y_L_fit = p[0] * syns_L + p[1]
        y_S_fit = sigmoid(syns_S, popt[0], popt[1], popt[2])
        
        y_L_error = rmse(y_L_fit, y_L)
        y_S_error = rmse(y_S_fit, y_S)
        
        # calculate total error, weighted by the number of data points
        error_tot = y_L_error * (t_s+1) + y_S_error * (len(syns) - t_s -1)
        
        if error_tot < min_error:
            
            min_error = error_tot
            p_best = p
            popt_best = popt
            best_ts = t_s
    
    return best_ts, p_best, popt_best
    
def si_converter(q_fit, FS_freq_data, i_max=1):
    """
    q_fit: quadratic fit of frequency as a function of current
    FS_freq_data: frequency data for frequency as a function of synapses
    i_max: maximum test current
    
    Determines the smallest error between the quad fit and the FS data to convert frequency to soma current
    """
    
    di = 0.001 # nA
    i_vec = np.arange(0, i_max, di)
    output_current = []
    
    f_poly = q_fit[0]*i_vec ** 2 + q_fit[1]*i_vec + q_fit[2]
    
    # Iterate through each frequency
    for f in range(len(FS_freq_data)):
        
        index = 0
        smallest_diff = 10
        
        # Iterate through the entire poly fit
        for i in range(len(f_poly)):
            
            # Determine the closest frequency match, then append the corresponding current value
            diff = abs(f_poly[i]-FS_freq_data[f])
            if diff < smallest_diff:
                smallest_diff = diff
                index = i

        output_current.append(i_vec[index])
    
    return output_current

def si_interpolater(fi_i, fi_f, fs_f):
    print(fi_f)
    fun = scipy.interpolate.interp1d(fi_f, fi_i, kind='linear', bounds_error=False, fill_value="extrapolate")
    return fun(fs_f)
            
def plot_fit(syns, y, p, popt, t_s):
    """
    syns: synapse vector
    y: y vector (frequency or current)
    p: linear fit parameters
    popt: sigmoid fit parameters
    t_s: transition synapse
    
    Plots curve fits onto xy data
    """
    
    syns_L = syns[0:t_s+1]
    syns_S = syns[:]
    
    
    nonlinear_fit = sigmoid(syns_S, popt[0], popt[1], popt[2])
    linear_fit = syns_S * p[0] + p[1]
    
    ratio = nonlinear_fit / linear_fit
    max_index = np.where(ratio == abs(max(ratio)))[0][0]

    max_ratio_syn = t_s + max_index
    
    print('Max Nonlinearity with Model:')
    print('Occurs at ', max_ratio_syn, ' synapses.')
    print('Linear Fit Prediction: ', linear_fit[max_index])
    print('Sigmoid Fit Prediction: ', nonlinear_fit[max_index])
    print('Actual Value: ', y[max_index+t_s+1])
    print('Nonlinear-Linear Ratio (Fit-Fit): ', y[max_index+t_s+1]/linear_fit[max_index])
    print('')
    
    ratio = y[:] / linear_fit
    max_index = np.where(ratio == abs(max(ratio)))[0][0]

    max_ratio_syn = t_s + max_index
    
    print('Max Nonlinearity with Data:')
    print('Occurs at ', max_ratio_syn, ' synapses.')
    print('Linear Fit Prediction: ', linear_fit[max_index])
    print('Sigmoid Fit Prediction: ', nonlinear_fit[max_index])
    print('Actual Value: ', y[max_index+t_s+1])
    print('Nonlinear-Linear Ratio (Data-Fit): ', nonlinear_fit[max_index]/linear_fit[max_index])
    print('')
    
    plt.plot(syns, i_converted)
    line_plot(syns_L, p)
    sigmoid_plot(syns_S, popt)
    line_plot(syns_S, p, True)
    plt.xlabel("# of Synapses")
    plt.ylabel("Dendritic Current at Soma (nA)")
    plt.title("IS Curve (With Fit)")
    plt.legend(["Data", "Line Fit", "Sigmoid Fit", "Linear Projection"])

    print("Transition Point: ", transition_syn)

    print("\nLine Fit:\n")
    print("m: ", p[0])
    print("b: ", p[1], "\n")

    print("Sigmoid Fit:\n")
    print("x0: ", popt[0])
    print("A: ", yS_max)
    print("B: ", popt[1])
    print("k: ", popt[2])
    
```

```python
# LOAD SIGMOID CURVE DATA

f = np.loadtxt("inhibitory-gamma/current-data/current-conductance-study/sinusoidal-current-with-noise-fs.csv", delimiter=",")
syns = np.array(range(0, len(f)))

f_current = np.loadtxt("inhibitory-gamma/current-data/current-conductance-study/sinusoidal-current-with-noise-fi.csv", delimiter=",")
inj_i = f_current[:, 0]
f_i = f_current[:, 1]
```

```python
inj_i = inj_i[0:45]
f_i = f_i[0:45]
```

```python
# FI PLOT

plt.plot(inj_i, f_i)
q = np.polyfit(inj_i, f_i, 2)

quad_fit = q[0] * inj_i ** 2 + q[1] * inj_i + q[2]
plt.plot(inj_i, quad_fit)

plt.xlabel("Injected Current (nA)")
plt.ylabel("Somatic Frequency (Hz)")
plt.title("FI Curve")
plt.legend(["Data", "Quad Fit"])
```

```python
# PRELIMINARY PLOT FOR ESTIMATION

plt.plot(syns, f)
plt.xlabel("# of Synapses")
plt.ylabel("Somatic Frequency (Hz)")
plt.title("FS Curve")
```

```python
# CURRENT AT SOMA VS. # OF SYNAPSES

i_converted = si_converter(q, f)
i_interp = si_interpolater(inj_i, f_i, f)
plt.plot(syns, i_interp)
plt.xlabel("# of Synapses")
plt.ylabel("Somatic Current (nA)")
plt.title("IS Curve")
```
